```python
import pandas as pd
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="Healthcare GeoLocator/1.0 (korutla.r@northeastern.edu)")

def get_location_info(place_name):
    # Adjust or remove skip terms as necessary
    skip_terms = ['<None>', '\xa0']  # Consider handling 'Airport' and 'Helipad' differently if needed
    
    if any(term in place_name for term in skip_terms):
        return None, None

    try:
        location = geolocator.geocode(place_name + ", USA", addressdetails=True)
        if location and location.raw.get('address'):
            address = location.raw['address']
            city = address.get('city', address.get('town', address.get('village')))
            county = address.get('county')
            return city, county
    except Exception as e:
        print(f"Error geocoding '{place_name}': {e}")
    return None, None

unique_hospitals = df2['Sending'].unique()
# Assuming 'unique_hospitals' contains all the names you want to geocode
hospital_locations = {}

for hospital in unique_hospitals:
    if hospital not in hospital_locations:
        city, county = get_location_info(hospital)
        if city and county:
            hospital_locations[hospital] = (city, county)
            time.sleep(1)  # Respect Nominatim's usage policy

# Update df2 with the obtained city and county information
df2['sender city'] = df2['Sending'].map(lambda x: hospital_locations.get(x, (None, None))[0])
df2['sender county'] = df2['Sending'].map(lambda x: hospital_locations.get(x, (None, None))[1])

# Display the first 5 rows to verify
df2.head()
```

```python
# Ensure 'tdate' is in datetime format
df2['tdate'] = pd.to_datetime(df2['tdate'])

# Combine 'tdate' with each time column to create full datetime columns
for col in ['disptime', 'enrtime', 'atstime', 'tratime', 'atdtime']:
    df2[col] = pd.to_datetime(df2['tdate'].dt.date.astype(str) + ' ' + df2[col].astype(str))
    # Adjust for events that end after midnight
    df2[col] = df2.apply(lambda row: row[col] + pd.Timedelta(days=1) if row[col] < row['tdate'] else row[col], axis=1)

# Calculate the time differences in minutes
df2['time_to_reach_patient_mins'] = (df2['atstime'] - df2['enrtime']).dt.total_seconds() / 60
df2['time_from_patient_to_hospital_mins'] = (df2['atdtime'] - df2['tratime']).dt.total_seconds() / 60

# Display the first 5 rows of the updated DataFrame
df2.head()
```
